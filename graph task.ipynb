```python
import networkx as nx
```

```python
import matplotlib.pyplot as plt
%matplotlib notebook
```

```python
graph=nx.DiGraph(Info="mark's details")
```

```python
graph.add_node("mark",pos=(0,0))
graph.add_node("cat",pos=(-2,6))
graph.add_node("student",pos=(2,-5))
graph.add_node("animal",pos=(1,6))
graph.add_node("spinoff",pos=(-5,-5))
graph.add_node("california",pos=(4,6))
graph.add_node("soccer",pos=(-5,2))
graph.add_node("sports",pos=(0,-8))
graph.add_node("CSU",pos=(5,-1))
pos=nx.get_node_attributes(graph,"pos")
```

```python
graph.add_edge("mark","cat",weight="has a")
graph.add_edge("mark","student",weight="is a")
graph.add_edge("cat","animal",weight="is a")
graph.add_edge("mark","soccer",weight="plays")
graph.add_edge("mark","california",weight="lives in")
graph.add_edge("mark","CSU",weight="studied in")
graph.add_edge("mark","sports",weight="is a")
graph.add_edge("mark","cat",weight="loves")
graph.add_edge("mark","spinoff",weight="club name is")
weight=nx.get_edge_attributes(graph,"weight")
```

```python
plt.figure()
nx.draw_networkx(graph,pos,font_weight="bold",node_size=2000,font_size=7)
nx.draw_networkx_edge_labels(graph,pos,edge_labels=weight)
```
